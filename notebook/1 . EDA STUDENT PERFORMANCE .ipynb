```python
for col in numeric_features:
        # Update numeric_features to exclude 'Competitor Pricing'
    numeric_features = [feature for feature in numeric_features if feature in df.columns]
    
    # Now perform the outlier handling
    for col in numeric_features:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        df[col] = df[col].clip(lower, upper)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df[col] = df[col].clip(lower, upper)
```

```python
df['Month'] = df['Date'].dt.to_period('M')
monthly_df = df.groupby('Month')[['Units Sold', 'Demand Forecast']].sum().reset_index()
monthly_df['Month'] = monthly_df['Month'].dt.to_timestamp()
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(x='Month', y='Units Sold', data=monthly_df, label='Actual Sales', marker='o')
sns.lineplot(x='Month', y='Demand Forecast', data=monthly_df, label='Forecasted Demand', marker='s')

plt.title('Actual Sales vs Forecasted Demand (Month-wise)', fontsize=14)
plt.xlabel('Month')
plt.ylabel('Units')
plt.legend()
plt.grid(True)
plt.show()
```
